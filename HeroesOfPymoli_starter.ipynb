```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
```

## Player Count

* Display the total number of players


```python
players = purchase_data.loc[:, ["Gender", "SN", "Age"]]
players = players.drop_duplicates()
total_players = players.count()[0]

pd.DataFrame({"Total Players": [total_players]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = (purchase_data["Item ID"].nunique())
avg_item_price = purchase_data["Price"].mean()
number_of_purchases = purchase_data["Price"].count()
total_price = purchase_data["Price"].sum()

summary = pd.DataFrame({"Number of Unique Items": item_count,"Average Price": [avg_item_price], "Number of Purchases": [number_of_purchases], 
                        "Total Price": [total_price]})

summary ["Number of Unique Items"] = summary ["Number of Unique Items"].map("{:,}".format)
summary ["Average Price"] = summary ["Average Price"].map("${:,.2f}".format)
summary ["Number of Purchases"] = summary ["Number of Purchases"].map("{:,}".format)
summary ["Total Price"] = summary ["Total Price"].map("${:,.2f}".format)
summary = summary.loc[:,["Unique Items", "Average Price", "Number of Purchases", "Total Price"]]

summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_total = players["Gender"].value_counts()
gender_total_percentage = gender_total / total_players * 100
gender_demographics = pd.DataFrame({"Total Count": gender_total, "Percentage of Players": gender_total_percentage})

gender_demographics = gender_demographics.round(2)

gender_demographics
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data.groupby(["Gender"]).count()["Price"].rename("Total Purchase Count")
avg_purchase_price = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
total_purchase_price = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Price")
avg_total_purchase_per_person = avg_purchase_price / purchase_count

summary_data = pd.DataFrame({"Total Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price, 
                             "Total Purchase Price": total_purchase_price, "Average Total Purchase Price Per Person": avg_total_purchase_per_person})

summary_data ["Average Purchase Price"] = summary_data ["Average Purchase Price"].map("${:,.2f}".format)
summary_data ["Total Purchase Price"] = summary_data ["Total Purchase Price"].map("${:,.2f}".format)
summary_data ["Average Total Purchase Price Per Person"] = summary_data ["Average Total Purchase Price Per Person"].map("${:,.2f}".format)

summary_data
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 99.99]
group_names = ["Less than 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "Over 40"]

players["Age Group"] = pd.cut(players["Age"], bins, labels=group_names)

age_group_totals = players["Age Group"].value_counts()
age_group_percents = age_group_totals / total_players * 100
age_demographics = pd.DataFrame({"Total Player Count Per Age Group": age_group_totals, "Percentage of Total Overall Players": age_group_percents})

age_demographics = age_demographics.round(2)

age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

purchase_count = purchase_data.groupby(["Age Group"]).count()["Price"].rename("Age Group Purchase Count")
avg_purchase_price = purchase_data.groupby(["Age Group"]).mean()["Price"].rename("Average Age Purchase Price")
total_purchase_price = purchase_data.groupby(["Age Group"]).sum()["Price"].rename("Total Age Group Purchase Price")
avg_total_purchase_price_per_person = avg_purchase_price / purchase_count

age_summary = pd.DataFrame({"Age Group Purchase Count": purchase_count, "Average Age Purchase Price": avg_purchase_price, 
                         "Total Age Group Purchase Price": total_purchase_price, "Average Total Purchase Price Per Person": avg_total_purchase_price_per_person})

age_summary ["Average Age Purchase Price"] = age_summary ["Average Age Purchase Price"].map("${:,.2f}".format)
age_summary ["Total Age Group Purchase Price"] = age_summary ["Total Age Group Purchase Price"].map("${:,.2f}".format)
age_summary ["Average Total Purchase Price Per Person"] = age_summary ["Average Total Purchase Price Per Person"].map("${:,.2f}".format)
age_summary = age_summary.loc[:, ["Age Group Purchase Count", "Average Age Purchase Price", "Total Age Group Purchase Price", "Average Total Purchase Price Per Person"]]

age_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")
avg_purchase_price = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
total_purchase_price = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purchase Price")

top_spender_summary = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price, "Total Purchase Price": total_purchase_price})

top_spender_summary ["Average Purchase Price"] = top_spender_summary ["Average Purchase Price"].map("${:,.2f}".format)
top_spender_summary ["Total Purchase Price"] = top_spender_summary ["Total Purchase Price"].map("${:,.2f}".format)
top_spender_summary = top_spender_summary.loc[:,["Purchase Count", "Average Purchase Price", "Total Purchase Price"]]

top_spender_summary.sort_values("Total Purchase Price", ascending=False).head(10)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular_data = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

purchase_count = most_popular_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Purchase Count")
item_price = most_popular_data.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Item Price")
total_purchase_price = most_popular_data.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Purchase Price")

most_popular_summary = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": item_price, "Total Purchase Price": total_purchase_price})

most_popular_summary ["Purchase Count"] = most_popular_summary["Purchase Count"].map("{:,}".format)
most_popular_summary["Item Price"] = most_popular_summary["Item Price"].map("${:,.2f}".format)
most_popular_summary["Total Purchase Price"] = most_popular_summary["Total Purchase Price"].map("${:,.2f}".format)
most_popular_summary = most_popular_summary.loc[:,["Purchase Count", "Item Price", "Total Purchase Price"]]

most_popular_summary.sort_values("Purchase Count", ascending=False).head(10)
```
